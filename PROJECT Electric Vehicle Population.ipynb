## 10 - Cleaning Data

```python
df["Postal Code"] = df["Postal Code"].replace('?', np.nan)
```

```python
mode5= df['Postal Code'].mode()
mode5 = mode5[0]
df['Postal Code'].replace( np.nan ,mode5, inplace = True)
```

```python
df["Postal Code"] = df["Postal Code"].astype('object')
```

```python
df["Electric Range"] = df["Electric Range"].replace('?', np.nan)
```

```python
df["Electric Range"] = df["Electric Range"].astype('float')
```

```python
df["DOL Vehicle ID"] = df["DOL Vehicle ID"].replace('?', np.nan)
```

```python
df["DOL Vehicle ID"] = df["DOL Vehicle ID"].astype('object')
```

```python
df["Legislative District"] = df["Legislative District"].replace('?', np.nan)
```

```python
mode11= df['2020 Census Tract'].mode()
mode11 = mode11[0]
df['2020 Census Tract'].replace( np.nan ,mode11, inplace = True)
```

```python
df["2020 Census Tract"] = df["2020 Census Tract"].astype('object')
```

```python
median1 = df['Electric Range'].median()
df['Electric Range'].replace( np.nan ,median1, inplace = True)
```

```python
mode1 = df['Electric Vehicle Type'].mode()
mode1_vehicle_type = mode1[0]
df['Electric Vehicle Type'].replace( np.nan ,mode1_vehicle_type, inplace = True)
```

```python
mode2 = df['Make'].mode()
mode2_Make = mode2[0]
df['Make'].replace( np.nan ,mode2_Make, inplace = True)
```

```python
mode3= df['County'].mode()
mode3 = mode3[0]
df['County'].replace( np.nan ,mode3, inplace = True)
```

```python
mode4= df['City'].mode()
mode4 = mode4[0]
df['City'].replace( np.nan ,mode4, inplace = True)
```

```python
mode6= df['Model'].mode()
mode6 = mode6[0]
df['Model'].replace( np.nan ,mode6, inplace = True)
```

```python
mode7= df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].mode()
mode7 = mode7[0]
df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].replace( np.nan ,mode7, inplace = True)
```

```python
mode8= df['Legislative District'].mode()
mode8 = mode8[0]
df['Legislative District'].replace( np.nan ,mode8, inplace = True)
```

```python
mode9= df['Vehicle Location'].mode()
mode9 = mode9[0]
df['Vehicle Location'].replace( np.nan ,mode9, inplace = True)
```

```python
mode10= df['Electric Utility'].mode()
mode10 = mode10[0]
df['Electric Utility'].replace( np.nan ,mode10, inplace = True)
```

```python
df["Model Year"] = df["Model Year"].astype('int64')
```

```python
#Remove outliers
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range    

for i in df.columns:
    if df[i].dtype !='object':
        lr,ur=remove_outlier(df[i])
        df[i]=np.where(df[i]>ur,ur,df[i])
        df[i]=np.where(df[i]<lr,lr,df[i])
```
